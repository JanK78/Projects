```python
#Plotting charts

import matplotlib.pyplot as plt

gender_counts = df['gender'].value_counts()

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Creating a bar chart
bars = axs[0].bar(gender_counts.index, gender_counts.values)
axs[0].set_xlabel('Gender')
axs[0].set_ylabel('Count')
axs[0].set_title('Gender Distribution')

# Create a pie chart
axs[1].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Gender Distribution')

# Adjusting spacing between subplots
plt.tight_layout()

# Displaying the plots
plt.show()
```

```python
# Creating a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# First subplot (Box Plot)
ax1.boxplot(df['age'], vert=False)
ax1.set_xlabel('Age')
ax1.set_title('Age Distribution')

# Second subplot (Histogram)
ax2.hist(df['age'], bins=20, edgecolor='k')
ax2.set_xlabel('Age')
ax2.set_ylabel('Frequency')
ax2.set_title('Age Distribution')

# Adjusting the layout to prevent overlapping titles
plt.tight_layout()

# Displaying the plots
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculating the percentage of cases with signs of mental illness
with_signs_mental_illness = df['signs_of_mental_illness'].sum()
without_signs_mental_illness = len(df) - with_signs_mental_illness

# Creating a pie chart for 'signs_of_mental_illness' vs 'all the rest'
labels = ['Signs of Mental Illness', 'No Signs of Mental Illness']
sizes = [with_signs_mental_illness, without_signs_mental_illness]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)  # Exploding the first slice (signs of mental illness)

fig, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

plt.title('Signs of Mental Illness vs Others')
plt.show()
```

```python
import pandas as pd

# Filtering the DataFrame to select only unarmed people
unarmed_df = df[df['armed'] == 'unarmed']

# Calculating the percentage of unarmed people
percentage_unarmed = (len(unarmed_df) / len(df)) * 100

print(f"Percentage of unarmed people: {percentage_unarmed:.2f}%")

# Filtering the DataFrame to select only people having a gun
gun_df = df[df['armed'].str.contains('gun', case=False, na=False)]

# Calculating the percentage of people armed with a gun
percentage_gun = (len(gun_df) / len(df)) * 100

print(f"Percentage of people armed with a gun: {percentage_gun:.2f}%")

percentage_other_object = 100 - percentage_unarmed - percentage_gun

print(f"Percentage of people with some other object: {percentage_other_object:.2f}%")
```

```python
# Creating a pie chart
labels = ['Unarmed', 'Gun', 'Other Object']
sizes = [percentage_unarmed, percentage_gun, percentage_other_object]
colors = ['#ff9999', '#66b3ff', '#99ff99']

fig, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')

plt.title('Percentage of People by Armament')
plt.show()
```

```python
crime_count_per_state = df['state'].value_counts().reset_index()

# Renaming the columns to be more descriptive
crime_count_per_state.columns = ['State', 'Incident Count']

# Displaying the result
print(crime_count_per_state)

# Creating a bar chart
plt.figure(figsize=(10, 6))
plt.bar(crime_count_per_state['State'], crime_count_per_state['Incident Count'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Incident Count')
plt.title('Incident Count per State')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Resampling the data by month and summing the incident counts
monthly_data = df.resample('M').sum()

# Finding the month(s) with the highest incident counts
highest_crime_months = monthly_data[monthly_data['id'] == monthly_data['id'].max()]

```

```python
import numpy as np

plt.figure(figsize=(12, 6))
plt.plot(monthly_data.index, monthly_data['id'], label='Monthly Incident Counts', marker='o')

# Fitting a linear regression line to the data
x = np.arange(len(monthly_data.index))
y = monthly_data['id'].values
coefficients = np.polyfit(x, y, 1)
trendline = np.poly1d(coefficients)

# Plotting the trendline
plt.plot(monthly_data.index, trendline(x), color='green', label='Trendline', linestyle='--')

# Continue with your existing code
plt.scatter(highest_crime_months.index, highest_crime_months['id'], color='red', label='Highest Crime Months', zorder=5)
plt.title('Monthly Incident Count Over Time and Trendline')
plt.xlabel('Date')
plt.ylabel('Incident Count')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
